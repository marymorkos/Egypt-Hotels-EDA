```python
summary_stats = data.describe()
print("Summary Statistics:\n", summary_stats)

missing_values = data.isnull().sum()
print("\nMissing Values:\n", missing_values)
```

```python
data = data.dropna(subset=['Overall Rating'])

data['Location Rating'].fillna(data['Location Rating'].median(), inplace=True)
```

```python
data['Total Rate (Lowest)'] = data['Total Rate (Lowest)'].replace('[EGP,]', '', regex=True).astype(float)
data['Rate per Night (Lowest)'] = data['Rate per Night (Lowest)'].replace('[EGP,]', '', regex=True).astype(float)
```

```python
data['Check-In Time'] = pd.to_datetime(data['Check-In Time'], errors='coerce')
data['Check-Out Time'] = pd.to_datetime(data['Check-Out Time'], errors='coerce')
```

```python
data['Amenities'] = data['Amenities'].fillna('')
amenities = data['Amenities'].str.split(',').explode()
```

```python
data.drop(columns=['Link'], inplace=True)
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Overall Rating'], bins=20, kde=True)
plt.title('Distribution of Overall Rating')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
data['Amenities'] = data['Amenities'].fillna('')
amenities = data['Amenities'].str.split(',').explode()  # Expand amenities into individual rows
amenities_counts = amenities.value_counts().head(10)    # Top 10 amenities
print(amenities_counts)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=amenities_counts.values, y=amenities_counts.index)
plt.title('Top 10 Amenities')
plt.xlabel('Frequency')
plt.ylabel('Amenity')
plt.show()
```

```python
data['Total Rate (Lowest)'] = data['Total Rate (Lowest)'].replace('[EGP,]', '', regex=True).astype(float)

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Location Rating', y='Total Rate (Lowest)', data=data)
plt.title('Location Rating vs Total Rate')
plt.xlabel('Location Rating')
plt.ylabel('Total Rate (Lowest)')
plt.show()
```

```python
import folium

# Define a map centered around the average latitude and longitude
map_center = [data['Latitude'].mean(), data['Longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=6)

# Filter hotels with a rating of 4.0 or above
high_rated_hotels = data[data['Overall Rating'] >= 4.0]

# Add markers for high-rated hotels
for _, row in high_rated_hotels.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"Hotel: {row['Name']}<br>Rating: {row['Overall Rating']}",
        icon=folium.Icon(color='blue' if row['Overall Rating'] >= 4.5 else 'green')
    ).add_to(m)

# Save map as HTML for presentation
m.save("High_Rated_Hotels_Map.html")
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point  # Make sure Point is imported
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from sklearn.cluster import KMeans

# Create geometry for spatial data using latitude and longitude
data['geometry'] = data.apply(lambda x: Point((x['Longitude'], x['Latitude'])), axis=1)
gdf = gpd.GeoDataFrame(data, geometry='geometry', crs="EPSG:4326")

# Filter to only high-rated hotels
high_rated_hotels_gdf = gdf[gdf['Overall Rating'] >= 4.0]

# Extract coordinates for clustering
coords = high_rated_hotels_gdf[['Latitude', 'Longitude']]

# Apply KMeans clustering
kmeans = KMeans(n_clusters=5, random_state=0).fit(coords)
high_rated_hotels_gdf['Cluster'] = kmeans.labels_

# Plot the clusters
fig, ax = plt.subplots(figsize=(10, 6))
high_rated_hotels_gdf.plot(column='Cluster', cmap='viridis', markersize=5, ax=ax, legend=True)
plt.title("Clusters of High-Rated Hotels by Region")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
